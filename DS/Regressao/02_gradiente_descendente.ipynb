```python
import numpy as np
import matplotlib.pyplot as plt

# Explicação inicial
print("O que é o Gradiente Descendente?")
print("- É um algoritmo usado para minimizar funções de erro ou custo ajustando os parâmetros do modelo.")
print("- Funciona iterativamente movendo-se na direção oposta ao gradiente (derivada) da função de custo.")
print("- Objetivo: Encontrar o ponto de menor erro possível (mínimo global ou local).\n")

# Função de custo (parabólica para facilitar)
def cost_function(x):
    return x**2 + 3*x + 2

# Derivada da função de custo (gradiente)
def gradient(x):
    return 2*x + 3

# Gradiente Descendente
def gradient_descent(learning_rate=0.1, iterations=10, start_point=0):
    x = start_point  # Ponto inicial
    history = []  # Para armazenar o progresso

    print(f"Executando o Gradiente Descendente:")
    print(f"- Taxa de aprendizado (learning_rate): {learning_rate}")
    print(f"- Número de iterações: {iterations}")
    print(f"- Ponto inicial: {x}\n")

    for i in range(iterations):
        cost = cost_function(x)
        grad = gradient(x)
        x = x - learning_rate * grad  # Atualização do parâmetro
        history.append((x, cost))  # Salvando o progresso
        print(f"Iteração {i+1}: x = {x:.4f}, Custo = {cost:.4f}, Gradiente = {grad:.4f}")

    return history

# Configurações do Gradiente Descendente
learning_rate = 0.1
iterations = 10
start_point = 5  # Escolha arbitrária para começar

# Executando o Gradiente Descendente
history = gradient_descent(learning_rate, iterations, start_point)

# Visualizando o progresso
x_vals = np.linspace(-6, 6, 100)  # Valores para a função de custo
y_vals = cost_function(x_vals)  # Calculando o custo para cada valor de x

# Plot
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label="Função de Custo", color="blue")
for i, (x, cost) in enumerate(history):
    plt.scatter(x, cost, color="red")
    plt.text(x, cost, f"  Iter {i+1}", fontsize=8)
plt.title("Gradiente Descendente: Convergência para o Mínimo")
plt.xlabel("x")
plt.ylabel("Custo")
plt.legend()
plt.grid(True)
plt.show()
```
