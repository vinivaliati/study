```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import validation_curve
from sklearn.linear_model import Ridge
from sklearn.datasets import make_regression

# Gerando dados sintticos para regresso
np.random.seed(42)
X, y = make_regression(n_samples=200, n_features=1, noise=15, random_state=42)

# Modelo de Regresso Ridge
model = Ridge()

# Hiperparmetro a ser avaliado: 'alpha' (regularizao)
param_range = np.logspace(-3, 3, 10)  # Valores de alpha variando de 0.001 a 1000 em escala logartmica

# Usando validation_curve para medir o desempenho do modelo
train_scores, test_scores = validation_curve(
    estimator=model,             # Modelo a ser avaliado
    X=X,                         # Dados de entrada
    y=y,                         # Target
    param_name="alpha",          # Hiperparmetro a variar
    param_range=param_range,     # Valores do hiperparmetro
    cv=5,                        # Validao cruzada com 5 folds
    scoring="r2",                # Mtrica: R
    n_jobs=-1                    # Usa todos os ncleos disponveis para acelerar
)

# Calculando a mdia e o desvio padro dos scores em cada fold
train_mean = np.mean(train_scores, axis=1)
train_std = np.std(train_scores, axis=1)
test_mean = np.mean(test_scores, axis=1)
test_std = np.std(test_scores, axis=1)

# Plotando a Validation Curve com reas
plt.figure(figsize=(8, 6))

# rea para os scores de treino
plt.fill_between(param_range, train_mean - train_std, train_mean + train_std, color="blue", alpha=0.3, label="Treino (com desvio padro)")

# rea para os scores de validao
plt.fill_between(param_range, test_mean - test_std, test_mean + test_std, color="orange", alpha=0.3, label="Validao (com desvio padro)")

# Linha central para facilitar a visualizao (opcional)
plt.plot(param_range, train_mean, color="blue", label="Mdia do Treino", linestyle="--")
plt.plot(param_range, test_mean, color="orange", label="Mdia da Validao", linestyle="--")

# Configurando o grfico
plt.xscale("log")  # Escala logartmica no eixo x (para visualizar melhor os valores de alpha)
plt.title("Validation Curve - Ridge Regression")
plt.xlabel("Alpha (Regularizao)")
plt.ylabel("R Score")
plt.legend(loc="best")
plt.grid(True)
plt.show()
```
