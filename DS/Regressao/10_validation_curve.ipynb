```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import validation_curve
from sklearn.linear_model import Ridge
from sklearn.datasets import make_regression

# Gerando dados sintéticos para regressão
np.random.seed(42)
X, y = make_regression(n_samples=200, n_features=1, noise=15, random_state=42)

# Modelo de Regressão Ridge
model = Ridge()

# Hiperparâmetro a ser avaliado: 'alpha' (regularização)
param_range = np.logspace(-3, 3, 10)  # Valores de alpha variando de 0.001 a 1000 em escala logarítmica

# Usando validation_curve para medir o desempenho do modelo
train_scores, test_scores = validation_curve(
    estimator=model,             # Modelo a ser avaliado
    X=X,                         # Dados de entrada
    y=y,                         # Target
    param_name="alpha",          # Hiperparâmetro a variar
    param_range=param_range,     # Valores do hiperparâmetro
    cv=5,                        # Validação cruzada com 5 folds
    scoring="r2",                # Métrica: R²
    n_jobs=-1                    # Usa todos os núcleos disponíveis para acelerar
)

# Calculando a média e o desvio padrão dos scores em cada fold
train_mean = np.mean(train_scores, axis=1)
train_std = np.std(train_scores, axis=1)
test_mean = np.mean(test_scores, axis=1)
test_std = np.std(test_scores, axis=1)

# Plotando a Validation Curve com áreas
plt.figure(figsize=(8, 6))

# Área para os scores de treino
plt.fill_between(param_range, train_mean - train_std, train_mean + train_std, color="blue", alpha=0.3, label="Treino (com desvio padrão)")

# Área para os scores de validação
plt.fill_between(param_range, test_mean - test_std, test_mean + test_std, color="orange", alpha=0.3, label="Validação (com desvio padrão)")

# Linha central para facilitar a visualização (opcional)
plt.plot(param_range, train_mean, color="blue", label="Média do Treino", linestyle="--")
plt.plot(param_range, test_mean, color="orange", label="Média da Validação", linestyle="--")

# Configurando o gráfico
plt.xscale("log")  # Escala logarítmica no eixo x (para visualizar melhor os valores de alpha)
plt.title("Validation Curve - Ridge Regression")
plt.xlabel("Alpha (Regularização)")
plt.ylabel("R² Score")
plt.legend(loc="best")
plt.grid(True)
plt.show()
```
